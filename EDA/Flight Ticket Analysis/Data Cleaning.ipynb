# DATA CLEANING

```python
final_data = final_data.copy()
```

#### Dropping Duplicates

```python
final_data = final_data.drop_duplicates(ignore_index = True)
```

#### Price column cleaning

```python
final_data["price"]
```

```python
final_data["price"] = final_data["price"].str.replace("INR","")
```

```python
final_data["price"] = final_data["price"].str.replace(",","")
```

```python
final_data["price"] = final_data["price"].astype(float)
```

```python
final_data["price"]
```

#### depature city column cleaning

```python
final_data["departure_city"].unique()
```

```python
final_data["departure_city"] = final_data["departure_city"].str.replace("DEL","Delhi").str.replace("CCU","Kolkata").str.replace("BLR","Banglore").str.replace("BOM","Mumbai").str.replace("MAA","Chennai")
```

```python
final_data["departure_city"]
```

```python
final_data["departure_city"].unique()
```

#### Arrival city column cleaning

```python
final_data["arrival_city"].unique()
```

```python
final_data["arrival_city"] = final_data["arrival_city"].str.replace("HYD","Hyderabad").str.replace("CCU","Kolkata").str.replace("BLR","Banglore").str.replace("BOM","Mumbai").str.replace("GAU","Guwahathi").str.replace("GOX","Goa")
```

#### travel date cleaning

```python
final_data["travel_date"]
```

```python
def add_year(travel_date):
    month = pd.to_datetime(travel_date, format = "%d %b").month
    if month == 12:
        return f"{travel_date} 2024"
    else:
        return f"{travel_date} 2025"
```

```python
final_data["travel_date"] = final_data["travel_date"].apply(add_year)
```

```python
final_data["travel_date"] = pd.to_datetime(final_data["travel_date"])
final_data["travel_date"]
```

```python
final_data
```

#### Booking Date cleaning 

```python
final_data["booking_date"] = final_data["booking_date"].apply(add_year)
```

```python
final_data["booking_date"] = pd.to_datetime(final_data["booking_date"])
final_data["booking_date"]
```

```python
final_data
```

```python
final_data.info()
```

### Calculating difference between travel_date and booking_date

```python
final_data["days_until_travel"] = (final_data["travel_date"] - final_data["booking_date"]).dt.days
```

final_data

```python
def booking_type(days_until_travel):
    if days_until_travel <= 7:
        return "Last-minute" 
    elif days_until_travel > 7 and days_until_travel <=30:
        return "Moderate"
    else:
        return "Advanced"
```

```python
final_data["booking_type"] = final_data["days_until_travel"].apply(booking_type)
```

```python
final_data
```

```python
holidays = ['2024-12-25', '2025-01-01', '2025-01-26']
holidays = pd.to_datetime(holidays)
def travel_type(travel_date):
    if travel_date in holidays:
        return 'Holiday'
    elif travel_date.weekday() >= 5:
        return 'Weekend'
    else:
        return 'Weekday'
```

```python
final_data["travel_type"] = final_data["travel_date"].apply(travel_type)
```

```python
final_data
```

#### Duration column cleaning

```python
final_data["duration"]
```

```python
def convert_to_minutes(duration):
    hours = int(duration.split('h')[0]) if 'h' in duration else 0
    minutes = int(duration.split('h')[-1].replace('m', '').strip()) if 'm' in duration else 0
    return hours * 60 + minutes
```

```python
final_data["duration"] = final_data["duration"].apply(convert_to_minutes)
```
