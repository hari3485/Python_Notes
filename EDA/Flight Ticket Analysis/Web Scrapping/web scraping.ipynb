### brands

```python
soup.find_all("div",class_="KzDlHZ")[0].text
```

```python
soup.find_all("div",class_="KzDlHZ")[15].text.split()[0]
```

```python
data = soup.find_all("div",class_="KzDlHZ")
```

```python
brands = []
for i in data:
    brands.append(i.text.split()[0])
```

```python
brands
```

```python
data[1].text.split("(")[0]
```

```python
models = []
for i in data:
    models.append(i.text.split("(")[0])
```

```python
data[1].text.split(",")[0].split("(")[1]
```

```python
colors = []
for i in data:
    colors.append(i.text.split(",")[0].split("(")[1])
```

```python
features = soup.find_all("li",class_="J+igdf")
```

```python
for i in features:
    print(i.text)
```

```python
features[6].text
```

```python
import re
```

```python
re.findall("(\d+)\s\w+\sRAM",features[6].text)[0]
```

```python
RAM = []
for i in features:
    a = i.text
    b = re.findall("(\d+)\s\w+\sRAM",a)
    if len(b)>0:
        RAM.append(b[0])
```

```python
features[0].text
```

```python
re.findall("(\d+)\s\w+\sROM",features[0].text)[0]
```

```python
ROM = []
for i in features:
    a = i.text
    b = re.findall("(\d+)\s\w+\sROM",a)
    if len(b)>0:
        ROM.append(b[0])
```

```python
ROM
```

```python
features[1].text
```

```python
re.findall("(\d+\.\d+)\sinch",features[1].text)[0]
```

```python
display = []
for i in features:
    a = i.text
    b = re.findall("(\d+\.\d+)\sinch",a)
    if len(b)>0:
        display.append(b[0])
```

```python
display
```

```python
features[3].text
```

```python
re.findall("(\d+)\smAh",features[3].text)[0]
```

```python
battery = []
for i in features:
    a = i.text
    b = re.findall("(\d+)\smAh",a)
    if len(b)>0:
        battery.append(b[0])
```

```python
battery
```

### Dataframe creation

```python
import pandas as pd
```

```python
d = {"Brand":brands,
    "Model":models,
    "RAM":RAM,
    "ROM":ROM,
    "Display":display,
    "Color":colors,
    "Battery":battery,
    "Price":prices}
```

```python
df = pd.DataFrame(d)
```
