```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://en.wikipedia.org/wiki/List_of_countries_and_territories_where_English_is_an_official_language"
page = requests.get(url)
```

```python
#200 - Success
400 - client side error
404 - Page not found
500 - server side error
501 - server side error
```

```python
page.text
```

```python
soup = BeautifulSoup(page.text)   # it converts string data into html
```

```python
soup
```

```python
soup.find_all("span",class_="mw-page-title-main")
```

```python
soup.find_all("span",class_="mw-page-title-main")[0]  # removing list
```

```python
soup.find_all("span",class_="mw-page-title-main")[0].text  # removing html tags
```
