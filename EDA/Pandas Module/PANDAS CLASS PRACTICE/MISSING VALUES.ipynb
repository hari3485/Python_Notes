# 4/10/2024

1. Outliers are values in data set that are very large or very small compared to majority values in the dataset 

2. Quartile deviation is defined as the difference between upper quartile (Q1) and loweer quartile (Q3)

3. Lower whisker = Q1 - 1.5(IQR) & Upper Whisker = Q3 + 1.5(IQR)

4. Outliers effects the model while training

```python
data = np.array([1,2,3,20,22,23,24,25,26,27,28,30,100,105,110])
data
```

```python
q1 = np.percentile(data, 25)
q2 = np.percentile(data, 50)
q3 = np.percentile(data , 75)
print(f"Quartiles :q1 : {q1},  q2 : {q2} ,  q3 : {q3}")
```

```python
iqr = q3 - q1
print(f"iqr : {iqr}")

lower_whisker = q1 - 1.5* iqr
upper_whisker = q3 + 1.5* iqr

print(f"lower_whisker : {lower_whisker}")
print(f"upper_whisker : {upper_whisker}")
```

```python
outliers = data[(data < lower_whisker) | (data > upper_whisker)]
outliers
```

```python
data.mean()
```

```python
np.median(data)
```

```python
a =np.array([20,22,23,24,25,26,27,28,30])
print(f"mean is : {np.mean(a)}")
print(f"median is : {np.median(a)}")
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (10,3))  #Figsize = (Width,Height)
plt.boxplot(data,vert = False)
```

```python
lower_outliers = data[(data < lower_whisker)]
upper_outliers = data[(data > upper_whisker)]
print(lower_outliers , lower_whisker,q1,q2,q3, upper_whisker,upper_outliers)
```

```python
jobs = pd.read_csv("pandas datasets/jobselection.csv")
jobs
```

```python
q1 = np.percentile(jobs["exam_marks"], 25)
q2 = np.percentile(jobs["exam_marks"], 50)
q3 = np.percentile(jobs["exam_marks"], 75)
print(f"Quartiles :q1 : {q1},  q2 : {q2} ,  q3 : {q3}")
```

```python
iqr = q3 - q1
print(f"iqr : {iqr}")

lower_whisker = q1 - 1.5* iqr
upper_whisker = q3 + 1.5* iqr

print(f"lower_whisker : {lower_whisker}")
print(f"upper_whisker : {upper_whisker}")
```

```python
outliers = jobs["exam_marks"][( jobs["exam_marks"]< lower_whisker) | (jobs["exam_marks"] > upper_whisker)]
outliers
```

```python
lower_outliers = jobs["exam_marks"][(jobs["exam_marks"] < lower_whisker)]
upper_outliers = jobs["exam_marks"][(jobs["exam_marks"] > upper_whisker)]
print(f"Upper outliers are : {upper_outliers}")
print(f"lower outliers are : {lower_outliers}")
```

```python
outliers.values
```

```python
plt.figure(figsize = (7,3))
jobs["exam_marks"].plot(kind = "box",vert = False)
plt.axvline(upper_whisker,color = "magenta",linestyle = "--",
           label=f"upper_whisker")
plt.axvline(lower_whisker,color = "blue",linestyle = "--",label = f"upper_whisker")
plt.legend()
```

```python
# Trimming is a technique to remove outliers from the dataframe
```
