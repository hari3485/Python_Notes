```python
import pandas as pd
titanic = pd.read_csv("pandas datasets/Titanic.csv")
titanic
```

#####  How can you filter out passengers who are younger than 18 and display their 'Name', 'Age', and 'Sex'?

```python
df = titanic.loc[titanic["Age"] < 18,["Name","Age","Sex"]]
df
```

#### How can you filter female passengers who paid more than 100 for the fare and show their 'Name' and 'Fare'?

```python
cond1 = titanic["Sex"] == "female"
cond2 = titanic["Fare"] > 100
titanic.loc[cond1 & cond2,["Name","Age"]]
```

#### How can you filter passengers who embarked from 'C' (Cherbourg) and are older than 40, displaying 'Name' and 'Age'?

```python
cond1 = titanic["Embarked"] == "C"
cond2 = titanic["Age"] > 40
titanic.loc[cond1 & cond2,["Name","Age"]]
```

#### How can you filter out male passengers whose fare was between 20 and 50, and display their 'Name', 'Age', and 'Fare'?


```python
cond1 = titanic["Sex"] == "male"
cond2 = titanic["Fare"].between(20,50)
titanic.loc[cond1 & cond2,["Name","Age","Fare"]]
```

```python
import numpy as np
data = {
    "id" : np.arange(1,11),
    "age" : [25, 45, 35, 28, 55, 60, 32, 42, 39, 50],
    "smoker" : [0,1,0, 0, 1, 0, 0, 1, 0, 1],   # 0:non-smoker,  1 - smoker
    "excercise_hours_per_week" : [3,1,4, 2, 0, 5, 2, 1, 3, 0],
 
    "cholestrol level" : [180, 240, 200, 195, 230, 220, 180, 210, 190, 245]   # Normal < 200, High >= 240
}
df_health=pd.DataFrame(data)
df_health
```

```python
def calculate_risk_score(row):
    score = 0
    score += (row["age"]-20)/10  # Adding points as age
    
    if row["smoker"]:
        score += 5
    score -= row["excercise_hours_per_week"]
    
    if row["cholestrol level"] >= 240:
        score += 4
    elif row["cholestrol level"] >= 200:
        score += 2
        
    return score
```

```python
df_health["risk_score"] = df_health.apply(calculate_risk_score,axis = 1)
df_health
```
