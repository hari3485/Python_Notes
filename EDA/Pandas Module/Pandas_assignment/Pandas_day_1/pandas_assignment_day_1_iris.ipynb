##### 2. Write a Pandas program to find the sum, mean, max, min value of 'SepalLengthCm' column of dataframe

```python
print("The mean of the SepalLengthCm is ",df["SepalLengthCm"].mean())
print("The sum of the SepalLengthCm column is ",df["SepalLengthCm"].sum())
print("The maximum value from the SepalLengthCm is ",df["SepalLengthCm"].max())
print("The minimum value  of the SepalLengthCm is ",df["SepalLengthCm"].min())
```
