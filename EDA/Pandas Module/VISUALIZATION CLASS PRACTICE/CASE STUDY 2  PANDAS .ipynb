# What is the average of sales price by country sales

```python
average_sales_price = pd.pivot_table(
                                     data_cleaned,
                                     index = "Description",
                                     columns = "Country",
                                     aggfunc = "mean",
                                     margins = True,
                                     margins_name = "Total").fillna(0).sort_values(by = "Total",
                                                                                    ascending = False)
```

```python
def analyze_product_sales(product_description , df):
    product_data = data_cleaned[data_cleaned["Description"] == product_description]
    
    
    # Calculate the total sales amount and quantity sold

    total_sales = product_data["TotalAmount"].sum()
    total_quantity = product_data["Quantity"].sum()

    # calculate sales performance across all countries
    country_perfomance = product_data.groupby("Country")[["TotalAmount","Quantity"]].sum().sort_values(by = "TotalAmount",ascending = False)


    return {"Total sales amount " : total_sales,
           "Total quantity sales " : total_quantity}, country_perfomance
```
