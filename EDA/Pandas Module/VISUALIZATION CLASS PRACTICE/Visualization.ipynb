```python
days = pd.Series(np.random.standard_normal(10) , index = range(10))
days
```

```python
temp = pd.Series(np.random.standard_normal(10) , index = range(10))
temp
```

```python
humid =  pd.Series(np.random.standard_normal(10) , index = range(10))
humid
```

```python
plt.style.use("seaborn-v0_8-dark-palette")
plt.plot(days , temp, color ="m",
         marker = "v" ,linestyle = "--" , label = "Temperature")
plt.plot(days,humid , color = "r",
        marker = "X", linestyle = "-.", label = "Humidity")

plt.title("INDIA TEMPERATURE", fontsize = 18)
plt.xlabel("Days", fontsize = 13,color = "b")
plt.ylabel("TEMP & Humidity", fontsize = 13, color ="g")
plt.legend(loc = "centre")

plt.savefig("Temp & Humidity.jpg", format = "jpej" , dpi = 300)
```

```python
country = ["uk","usa","india","sa","nz","aus"]
gold = np.random.randint(10,50,(len(country,)))
silver =  np.random.randint(10,50,(6,))
bronze =  np.random.randint(10,50,(6,))
```

```python
medals = {"Countries" : country,
         "Gold" : gold,
         "Silver" : silver,
         "Bronze" : bronze
         }
medals = pd.DataFrame(medals)
medals.set_index("Countries",inplace = True)
medals
```

```python
# STACKED BAR GRAPH
plt.figure(figsize=(10,6))
medals.plot(kind = "barh",stacked = True ,color = ["gold","silver","brown"])
```

```python
plt.figure(figsize=(10,6))
medals.plot(kind = "barh",stacked = True ,color = ["gold","silver","brown"])

for index,value  in enumerate(medals.values):
    for i in range(len(value)):
        plt.text(sum(value[:i+1]) - value[i]/2 , index , int(value[i]),va = "center",ha = "center")
        
# Positioning the sum and returning at center of bar chart
# To get labels of y axis and y axis
plt.ylabel("countries")
plt.xlabel("medals count")
plt.legend(["gold","silver","bronze"] , loc = "best")
# ADDING Labels to each section of bars
```

# SUBPLOTS

```python
x = list("MUNNU") #in character i took repeted letters so i got only 3 col in output
y = np.random.randint(10,50,(5,))
plt.figure(figsize = (10,8))
# Creating subplots in a plo

# First subplot
plt.subplot(2,2,1)
plt.plot(x,y, marker  = "d")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Linear plot")


# Second subplot
plt.subplot(2,2,2)
plt.pie(y , autopct = "%0.2f%%")
plt.title("Pie chart")


# Third subplot
plt.subplot(2,2,3)
plt.bar(x,y)


#Fourthh subplot
plt.subplot(2,2,4)
plt.barh(x,y)

plt.suptitle("Multiple plots in a single canvas(figure)" , fontsize = 12)
print(x,y)
```

```python
x = list("HARIC") #in character i took unique letters so i got all col in output
y = np.random.randint(10,50,(5,))
plt.figure(figsize = (10,8))
# Creating subplots in a plo

# First subplot
plt.subplot(2,2,1)
plt.plot(x,y, marker  = "d")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Linear plot")


# Second subplot
plt.subplot(2,2,2)
plt.pie(y , autopct = "%0.2f%%" , labels = x, colors = plt.style.use("seaborn-v0_8-dark"))
plt.title("Pie chart")


# Third subplot
plt.subplot(2,2,3)
plt.bar(x,y)


#Fourthh subplot
plt.subplot(2,2,4)
plt.barh(x,y)

plt.suptitle("Multiple plots in a single canvas(figure)" , fontsize = 12)
print(x,y)
```
