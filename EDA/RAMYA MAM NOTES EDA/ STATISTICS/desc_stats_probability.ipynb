# Probability

Probability can be defined as the ratio of the number of favorable outcomes to the total number of outcomes of an exprimentr. For an experiment having 'n' number of outcomes, the number of favorable outcomes can be denoted by x. The formula to calculate the probability of an event is as follows.

Probability(Event) = Favorable Outcomes/Total Outcomes = x/n


# Terminology of Probability Theory



## 1. Experiment: 

A trial or an operation conducted to produce an outcome is called an experiment.

## 2. Sample Space: 

All the possible outcomes of an experiment together constitute a sample space. For example, the sample space of tossing a coin is {head, tail},
in throwing a unbiased dice (regular dice)-- sample space -- {1,2,3,4,5,6}

## 3.Favorable Outcome: 

An event that has produced the desired result or expected event is called a favorable outcome. For example, when we roll two dice, the possible/favorable outcomes of getting the sum of numbers on the two dice as 4 are (1,3), (2,2), and (3,1).


## 4.Random Experiment: 

An experiment that has a well-defined set of outcomes is called a random experiment. For example, when we toss a coin, we know that we would get ahead or tail, but we are not sure which one will appear.




## 5. Trial: 

A trial denotes doing a random experiment.


## 6.Event: 

Its the Subset of Sample Sapce. The total number of outcomes of a random experiment is called an event.

**EG:**

- RANDOM EXPERIMENT/ trail: throwing in the dice roll.

- event : getting prime numbers from it

- SAMPLE SPACE FOR DICE ROLL : {1,2,3,4,5,6}

- EVENT : {2,3,5}

- P(EVENT) = (NUMBER OF FAVOURABLE OUTCOMES FOR AN EVENT) / (SAMPLE SPACE OF RANDOM EXP) = |E| / 6 = 3/6



## 7.Equally Likely Events:

Events that have the same chances or probability of occurring are called equally likely events. The outcome of one event is independent of the other. For example, when we toss a coin, there are equal chances of getting a head or a tail.

## 8.Exhaustive Events: 

There are six possible outcomes when rolling a die which is {1, 2, 3, 4, 5, 6}. Now, if we roll a die, one of these six outcomes will definitely occur. Hence, all these six outcomes are exhaustive events. When the set of all outcomes of an event is equal to the sample space, we call it an exhaustive event.

# 1. Dependent Events?
For events to be considered dependent, one must have an influence over how probable another is. In other words, a dependent event can only occur if another event occurs first.


- If you park your vehicle illegally, you’re more likely to get a parking ticket.
- You must buy a lottery ticket to have a chance at winning; your odds of winning are increased if you buy more than one ticket.
- Committing a serious crime – such as breaking into someone’s home – increases your odds of getting caught and going to jail.

# 2. What are Independent Events?
An event is deemed independent when it isn’t connected to another event, or its probability of happening, or conversely, of not happening.

- Taking an Uber ride and getting a free meal at your favorite restaurant
- Winning a card game and running out of bread
- Finding a dollar on the street and buying a lottery ticket; finding a dollar isn’t dictated by buying a lottery ticket, nor does buying the ticket increase your chances of finding a dollar
- Growing the perfect tomato and owning a cat

# Random Variable

A random variable, also known as a stochastic variable, functions in the realm of real values within the entire sample space of an experiment. Imagine the domain as the pool of all possible values applicable to the function. Just as a function processes its domain, a random variable operates on its domain (the experiment’s sample space) and assigns a real value to each event or outcome.

1. Discrete Random Variables

Discrete random variables are random variables, whose range is a countable set. A countable set can be either a finite set or a countably infinite set. For instance, in the above example, X is a discrete variable as its range is a finite set ({0, 1, 2}).

2. Continuous Random Variables


Continuous random variables, on the contrary, have a range in the forms of some interval, bounded or unbounded, of the real line. E.g., Let Y be a random variable that is equal to the height of different people in a given population set. Since the people can have different measures of height (not limited to just natural numbers or any countable set), Y is a continuous variable.

3. Mixed Random Variables:

mixed random variables are ones that are a mixture of both continuous and discrete variables.
