A sample is the specific group that you will collect data from

```python
import numpy as np
a=np.array([8,109,62,57,61,3,58,60,62,9,110,61,57,122,58])
print(np.mean(a))
print(np.var(a))
#
```

```python
a=np.array([8,109,62,57,61,3,58,60,62,9,110,61,57,122,58])
n= a.size
s=0
m= np.mean(a)
for i in a:
    s= s+ ((i-m)**2)

print(s/n)
print(s/(n-1))
```
