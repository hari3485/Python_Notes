```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Number of samples per experiment
n_values = [1, 2, 10, 20, 30, 50, 100, 200]

# Number of experiments (or sets of random values)
n_experiments = 1_00_000

plt.figure(figsize= (18,12))

for i, n in enumerate(n_values):
    # Step 1: Generate random values from different distributions
    
    samples_pareto = np.random.pareto(a=5, size=(n_experiments, n)) + 1  # 'a' is shape
    samples_poisson = np.random.poisson(3, (n_experiments, n))  # lambda = 3
     
    # Step 2: Sum the values to stimulate the combined effect for each distribution  
    # sums_uniform = np.sum(samples_uniform, axis=1)
    sums_pareto = np.sum(samples_pareto, axis=1)
    sums_poisson = np.sum(samples_poisson, axis=1)
    combined_sum = sums_pareto + sums_poisson
    
    # Step 3: Plot the histogram of combined sums
    plt.subplot(3, 4, i+1)
    plt.hist(combined_sum, bins=50, density=True, alpha=0.6, color='blue', label= 'Combined Sum')
    plt.title(f"Sum of {n} Random Values")
    plt.legend()
```
