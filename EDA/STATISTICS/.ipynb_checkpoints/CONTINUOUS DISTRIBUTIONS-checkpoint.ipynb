# Normal Distribution

- In normal distribution __mean__ decides the `centre of distribution`
**variance** describes the `spread of distribution`.

- It is symetrical distribution where `mean = median = mode`



```python
x = np.array([1,2,2,3,3,4,4,5])
mean_x = np.mean(x)
std_x = np.std(x)
mean_x , std_x
```

```python
Z = (x - mean_x)/(std_x)
```

```python
# Here mean is "0" and std is "1" in Standard normal distribution

np.mean(Z),np.std(Z) 
```

- Practical Example
- Suppose you have a normal distribution of exam scores with a mean of 70 and a standard deviation of 10.

- PDF: To find the density around a score of 80, you'd look at the PDF at 80, which tells you how likely scores are to cluster around that value.
- CDF: If you want the probability that a student scored 80 or less, you'd use the CDF at 80, which gives the probability of scoring below or equal to 80.

```python
x_val = np.linspace(mean - 4*std, mean+4*std,1000)

pdf_values = stats.norm.pdf(x_val,mean,std)
```

```python
plt.plot(x_val , pdf_values , label = "pdf" , color = "blue")
plt.fill_between(x_val , pdf_values,
                where = (x_val <= 40),
                color = "m",
                alpha = 0.5)
plt.axvline(40, color = "blue",linestyle = "--" , label = "P(X<=40)")
#give 3sigma limits and observe the change
```

4 The weights of a population of adult men are normally distributed with a mean of 70 kg and a standard deviation of 8 kg. What is the probability that a randomly selected man weighs between 65 kg and 75 kg?


- Here there are two limits are given so to the weights between them frst we have to find p(x<=75) we get all weight less than equal to 75 then find p(x<=65) we get all weights less than equal to 65 now subtract p(x<=75) - p(x<=65) we get the weights which are in between 65 and 75

```python
mu = 70
sig = 8
p_75 = stats.norm.cdf(75,mu,sig)
p_65 = stats.norm.cdf(65,mu,sig)
print(p_75,p_65)
p_75 - p_65
```

- So, there is approximately  46% probability that weights will fall between 65 and 75.

```python
mu = 70
sig = 8

x_val = np.linspace(mu - 4*sig, mu+4*sig,1000)

pdf_values = stats.norm.pdf(x_val,mu,sig)
```

```python
plt.plot(x_val , pdf_values , label = "pdf" , color = "blue")
plt.fill_between(x_val , pdf_values,
                where = ((x_val >= 65) & (x_val <= 75)),
                color = "m",
                alpha = 0.5)
plt.axvline(65, color = "k",linestyle = "--" , label = "P(X>=65)")
plt.axvline(75, color = "g",linestyle = "--" , label = "P(X<=75)")
plt.ylim(0,0.055)
plt.legend()
#give 3sigma limits and observe the change
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Set parameters for the binomial distribution
n = 10      # number of trials
p = 0.9     # probability of success
size = 1000 # number of random variates

# Generate a binomial distribution
data = np.random.binomial(n, p, size)

# Convert the data to a Pandas DataFrame
df = pd.DataFrame(data, columns=['binomial_distribution'])

# Plot the distribution
df['binomial_distribution'].value_counts().sort_index().plot(kind='bar', color='skyblue')
plt.xlabel('Number of Successes')
plt.ylabel('Frequency')
plt.title('Binomial Distribution (n=10, p=0.9)')
plt.show()
```

```python
a = 10
b = 21

# Create a uniform random variable 
rv = uniform(loc = a , scale = b-a)

# Generate x values
x = np.linspace(a,b,1000)

# calculate pdf,cdf 
pdf_val = rv.pdf(x)
cdf_val = rv.cdf(x)
pdf_val,cdf_val
```

```python
plt.figure(figsize = (6, 3))

plt.vlines(x, 0, pdf_val, colors='orange', lw=2, zorder=1)

plt.title(f'Uniform Distribution PDF (a={a}, b={b})')

plt.xlabel('Value')

plt.xlim(9,22)

plt.ylim(0,0.3)

plt.ylabel('Density')

plt.show()
```
