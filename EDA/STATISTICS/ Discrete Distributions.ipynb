1. In a clinical trial a new drug has a 80% chance of curing a patient. Out of 20 patients 

```python
import scipy.stats as stats
import numpy as np

n = 20 # Number of independent bernoulli trials 
p = 0.8 # probability of sucess of single event 


p_18 = stats.binom.pmf(18, n, p)# P(X=18)
p_19 = stats.binom.pmf(19, n, p)# P(X=19)
p_20 = stats.binom.pmf(20, n, p)# P(X=20)
print(p_18, p_19, p_20)
p_atleast_18 = p_18+p_19+p_20 # P(X>=18) = P(X=18)+P(X=19)+P(X=20)
p_atleast_18
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Set parameters for the binomial distribution
n = 10      # number of trials
p = 0.9     # probability of success
size = 1000 # number of random variates

# Generate a binomial distribution
data = np.random.binomial(n, p, size)

# Convert the data to a Pandas DataFrame
df = pd.DataFrame(data, columns=['binomial_distribution'])

# Plot the distribution
df['binomial_distribution'].value_counts().sort_index().plot(kind='bar', color='skyblue')
plt.xlabel('Number of Successes')
plt.ylabel('Frequency')
plt.title('Binomial Distribution (n=10, p=0.9)')
plt.show()
```

```python
# import necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import bernoulli,binom,poisson,uniform,expon,norm,lognorm,pareto
```

```python
# paramteres for brnoulli distn
p = 0.7 # probability of sucess

# create a bernoulli random variable
x = [0,1]
rv = bernoulli(p)

pmf_val = rv.pmf(x)
cdf_val = rv.cdf(x)
pmf_val,cdf_val
```

-    fx(0) = p(x<=0)
          = p(x = 0)
          = 0.3

 -   fx(1) = p(x<=1)
          = p(x=0)+p(x=1)
          = 0.3 + 0.7
          = 1

```python
# lollipop plot
plt.figure(figsize = (4,3))
sns.scatterplot(x = x, y = pmf_val, s= 100,color = "m",zorder = 2)
plt.vlines (x, 0, pmf_val, colors='blue', lw=2, zorder=1)

plt.title(f'Bernoulli Distribution PMF (p={p})')

plt.xlabel('x')

plt.xlim(-0.5, 2)

plt.ylim(0,1)

plt.ylabel('Probability')

plt.show()
```

- z order in `scatterplot  as a`
- z order in `line as b`
- when we give b value greater than a val it returns some part of line comes over the circle

```python
# Step chart for CDF

plt.figure(figsize=(4, 3))

plt.step(x, cdf_val, where='post', color='green', lw=2)

plt.scatter(x, cdf_val, color='green', s=100, zorder=2)



plt.title(f'Bernoulli Distribution CDF (p={p})')

plt.xlabel('x')

plt.xlim(-0.5, 2)

plt.ylim(0.2,1.2)

plt.ylabel('Cumulative Probability')

plt.show()
```
