The __init__() Function

The examples above are classes and objects in their simplest form, and are not really useful in real life applications.

To understand the meaning of classes we have to understand the built-in __init__() function.

All classes have a function called __init__(), which is always executed when the class is being initiated.

Use the __init__() function to assign values to object properties, or other operations that are necessary to do when the object is being created:

**Example**

Create a class named Person, use the __init__() function to assign values for name and age:

```python
class human:
    def __init__(self,name,gender,age):
        self.name=name
        self.gender=gender
        self.age=age
        
obj1=human("ramya","female",16)
print(obj1.name,obj1.age)

tom = human("Tom cruise","male",62)
print(f"the name is {tom.name} and the age is {tom.age} and gender is {tom.gender}")
```

```python
class Person:
    def __init__(self, name, age):  # properties which can be used inside class need to be defined here
        self.name = name  # assigning the properties to the names
        self.age = age
        
o1= Person("ramya","female")
print(o1.age)

o1.age=30
print(o1.age)
```

```python
class human:
    def __init__(self,n,g,a):
        self.name=n
        self.gender=g
        self.age=a
        
obj1=human("ramya","female",16)
print(obj1.name,obj1.age)

obj2=human("ankit","male",20)
print(obj2.gender)
```

```python
class human:
    def __init__(self,n,g,a):
        self.name=n
        self.gender=g
        self.age=a
        print(id(self))
        
obj1=human("ramya","female",16)
print(id(obj1))

obj2=human("ramya","female",16)
print(id(obj2))
```
