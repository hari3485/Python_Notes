#### Creating a Vehicle Class

```python
class Vehicle:
    def __init__(self, type, no_of_wheel, fuel_type, company):   ### PARAMETERISED CONSTRUCTOR
        self.type = type
        self.wheels = no_of_wheel
        self.fuel_type = fuel_type
        self.Company = company

    def view_details(self):
        print("Vehicle type: {}\nWheels: {}\nFuel: {}\nCompany: {}".format(self.type,
                                                                      self.wheels,
                                                                      self.fuel_type,
                                                                      self.Company))
```

```python
car = Vehicle()
```

```python
car = Vehicle("Car", 4, "Deisel", "TATA")   ### PARAMETERISED CONSTRUCTOR
moped = Vehicle("Luna", 2, "Pedal", "TVS")
auto = Vehicle("Auto", 3, "CNG", "TVS")
```

```python
auto.view_details()
```

```python
id(auto.Company), id(moped.Company)
```

##### GETTERS and SETTERS

```python
class Student:
    def __init__(self, name, gender, roll_no, branch):
        self.name = name
        self.gender = gender
        self.roll = int(roll_no)
        self.branch = branch
    ### GETTER --- Fetch values
    def view_stud_details(self):
        print("Name: {}\nGender: {}\nRoll No: {}\nBranch : {}".format(self.name,
                                                                      self.gender,
                                                                      self.roll,
                                                                      self.branch
                                                                     ))
    ### SETTER --- Setting a new Value
    def update_branch(self, updated_branch):
        self.branch = updated_branch
        print(F"{self.name}, Your branch has been updated SUCCESSFULLY to {self.branch}...")
```

```python
stud1 = Student("Ganesh", "M", 101, "DS")
stud2 = Student("Bhaskar", "M", 102, "DA")
stud3 = Student("Priyanka", "F", 103, "DS")
```

## Types of variables
1. Instance Variable
    - If you create any variable inside the method, those are we call as instace variable.
    - You should use object to call any instace variable
2. Class Variable or Static Variable
    - If you create any variable outside the method and inside the class, those variables we called as class variables or static variable
    - In order to call class variables. There are 2 different ways
        1. By using object name
        2. By using class name

```python
class Student:
    institite = "Innomatics"
    def __init__(self, name, gender, roll_no, branch):
        self.name = name           ##### INSTANCE variable
        self.gender = gender
        self.roll = int(roll_no)
        self.branch = branch
    ### GETTER --- Fetch values
    def view_stud_details(self):
        print("Name: {}\nGender: {}\nRoll No: {}\nBranch : {}".format(self.name,
                                                                      self.gender,
                                                                      self.roll,
                                                                      self.branch
                                                                     ))
    ### SETTER --- Setting a new Value
    def update_branch(self, updated_branch):
        self.branch = updated_branch
        print(F"{self.name}, Your branch has been updated SUCCESSFULLY to {self.branch}...")
```
